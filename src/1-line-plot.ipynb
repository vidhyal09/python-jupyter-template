### Exercise 1: Line Plot Customization

```python
# TODO: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np

# Function to check if a list contains only numeric values
def validate_input(x):
    if not x:  # Check if the list is empty
        raise ValueError("The input list is empty.")
    if not all(isinstance(i, (int, float)) for i in x):  # Ensure all values are numeric
        raise ValueError("All values in the input list must be numeric.")
    return True

# Task: Create a line plot of y = x^2
x = [2, 4, 6, 8, 9, 10]  # x values ranging from 0 to 10

try:
    # Validate the input
    validate_input(x)
    
    # Compute y values as the square of x
    y = [i**2 for i in x]

    # Create the plot
    plt.plot(x, y)

    # Customize the plot
    plt.title('Line Plot of y = x^2')  # Add a title
    plt.xlabel('x')  # Label for x-axis
    plt.ylabel('y = x^2')  # Label for y-axis
    plt.grid(True)  # Add grid

    # Show the plot
    plt.show()

except ValueError as e:
    print(f"Error: {e}")
```
